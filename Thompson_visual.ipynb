```python
import random
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt
```

```python
prob = [0.4, 0.5]
n_versions = len(prob)

def getUserResponse(i):
    """
    Simulates user behaviour according to the probability in variable prob
    Returns True or False 
    """
    return prob[i]>random.uniform(0,1)
```

```python
def show_figure():
    x = [i/1000.0 for i in range(0,1001)]
    y1 = beta.pdf(x, hit[0], no_hit[0])
    y2 = beta.pdf(x, hit[1], no_hit[1])
    
    n = sum(hit) + sum(no_hit) - 4
    if n == 0:
        title_str = "Thompson sampling before any visits"
    elif n == 1:
        title_str = "Thompson sampling after 1 visit"
    else:
        title_str = "Thompson sampling after {} visits".format(n)
    
    plt.figure(figsize=(8,3))
    plt.plot(x,y1, color="black", linestyle="dashed", label="Low converting version", hold=True)
    plt.plot(x,y2, color="black", linestyle="dotted", label="High converting version")
    plt.legend(loc='best')
    plt.xlabel("Estimated convertion ratio")
    plt.ylabel("P ( Conversion | Data )")
    plt.title(title_str)
    plt.show()
    plt.close()
```

```python
hit    = np.array([1] * n_versions)
no_hit = np.array([1] * n_versions)

# Before any vists
show_figure()

# After first
do_thompson_step()
show_figure()

# After second
do_thompson_step()
show_figure()

# After third
do_thompson_step()
show_figure()

# After fourth
do_thompson_step()
show_figure()

# After ten
for i in range(6):
    do_thompson_step()
show_figure()

# After 100
for i in range(90):
    do_thompson_step()
show_figure()

# After 1000
for i in range(900):
    do_thompson_step()
show_figure()

print("Total visitors version 0: {}, buyers: {}".format(str(hit[0]+no_hit[0]),str(hit[0])))
print("Total visitors version 1: {}, buyers: {}".format(str(hit[1]+no_hit[1]),str(hit[1])))
```
