```python
# Import dependencies
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set up variables

prob = [0.3, 0.33, 0.25, 0, 0.05, 0.1]
n_versions = len(prob)

def getUserResponse(i):
    return prob[i]>random.uniform(0,1)

# Params for e-Greedy
eps = 0.2
```

```python
# Define e-Greedy
def eGreedy(n_versions, n_runs, start_eps, eps_decay=1):
    hit    = np.array([1.0] * n_versions)
    no_hit = np.array([1.0] * n_versions)
    belief = np.array([1.0/n_versions] * n_versions)
    eps = start_eps
    
    for run in range(n_runs):
        eps = eps*eps_decay
        
        # Choose which version to show
        if random.uniform(0,1) > eps:
            chosen_version = np.argmax(belief) # Choose best according to current belief
        else:
            chosen_version = random.randint(0,n_versions-1) # Explore
    
        # Compute user response
        resp = getUserResponse(chosen_version)
    
        # Update information
        if resp:
            hit[chosen_version] = hit[chosen_version] + 1
        else:
            no_hit[chosen_version] = no_hit[chosen_version] + 1
        belief = hit / (hit+no_hit)

    # Remove priors
    hit = hit - 1
    no_hit = no_hit - 1
    
    return hit, no_hit, belief
```

```python
# Define Thompson sampling
def thompson(n_versions, n_runs):
    hit    = np.array([1.0] * n_versions)
    no_hit = np.array([1.0] * n_versions)

    for run in range(n_runs):
        # Choose which version to show
        belief = np.zeros(n_versions)
        for i in range(n_versions):
            belief[i] = np.random.beta(hit[i],no_hit[i])
        chosen_version = np.argmax(belief) # Choose best according to current belief
    
        # Compute user response
        resp = getUserResponse(chosen_version)
    
        # Update information
        if resp:
            hit[chosen_version] = hit[chosen_version] + 1
        else:
            no_hit[chosen_version] = no_hit[chosen_version] + 1
    
    # Compute final belief
    belief = np.zeros(n_versions)
    for i in range(n_versions):
        belief[i] = np.random.beta(hit[i],no_hit[i])
    
    # Remove priors
    hit = hit - 1
    no_hit = no_hit - 1
    return hit, no_hit, belief, (hit / (hit+no_hit))
```

```python
# Compute results
batch_meta = 1000
meta_runs = [10]*batch_meta + [20]*batch_meta + [50]*batch_meta + [100]*batch_meta + [200]*batch_meta + [500]*batch_meta + [1000]*batch_meta + [2000]*batch_meta + [5000]*batch_meta + [10000]*batch_meta + [20000]*batch_meta + [50000]*batch_meta + [100000]*batch_meta
# + [100000]*batch_meta # + [1000000]*batch_meta

# It would probably be prettier to put these in a Dict.
hit_random      = []
hit_eGreedy     = []
hit_eGreedyD    = []
hit_thompson    = []
no_hit_random   = []
no_hit_eGreedy  = []
no_hit_eGreedyD = []
no_hit_thompson = []

result_random   = []
result_eGreedy  = []
result_eGreedyD = []
result_thompson = []
belief_random   = []
belief_eGreedy  = []
belief_eGreedyD = []
belief_thompson = []
belief2_thompson = []

for runs in meta_runs:
    # random
    hit, no_hit, belief = eGreedy(n_versions, runs, 1)
    result_random.append( np.sum(hit) / (np.sum(hit) + np.sum(no_hit)) )
    belief_random.append(belief)
    hit_random.append(hit)
    no_hit_random.append(no_hit)
    
    # e-Greedy
    hit, no_hit, belief = eGreedy(n_versions, runs, eps)
    result_eGreedy.append( np.sum(hit) / (np.sum(hit) + np.sum(no_hit)) )
    belief_eGreedy.append(belief)
    hit_eGreedy.append(hit)
    no_hit_eGreedy.append(no_hit)
    
    # e-Greedy decreasing
    hit, no_hit, belief = eGreedy(n_versions, runs, eps, eps_decay=0.999)
    result_eGreedyD.append( np.sum(hit) / (np.sum(hit) + np.sum(no_hit)) )
    belief_eGreedyD.append(belief)
    hit_eGreedyD.append(hit)
    no_hit_eGreedyD.append(no_hit)
    
    # Thompson sampling
    hit, no_hit, belief, belief2 = thompson(n_versions, runs)
    result_thompson.append( np.sum(hit) / (np.sum(hit) + np.sum(no_hit)) )
    belief_thompson.append(belief)
    belief2_thompson.append(belief2)
    hit_thompson.append(hit)
    no_hit_thompson.append(no_hit)
```

```python
# Compute average amount of hits
unique_meta_runs = sort(list(set(meta_runs)))
averages_r   = np.zeros([len(unique_meta_runs)])
averages_eG  = np.zeros([len(unique_meta_runs)])
averages_eGD = np.zeros([len(unique_meta_runs)])
averages_ts  = np.zeros([len(unique_meta_runs)])

for run_size, result_r, result_eG, result_eGD, result_ts in zip(meta_runs, result_random, result_eGreedy, result_eGreedyD, result_thompson):
    i = np.where(unique_meta_runs == run_size)[0][0]
    averages_r[i] = averages_r[i] + result_r
    averages_eG[i] = averages_eG[i] + result_eG
    averages_eGD[i] = averages_eGD[i] + result_eGD
    averages_ts[i] = averages_ts[i] + result_ts
    
averages_r  = averages_r / batch_meta
averages_eG  = averages_eG / batch_meta
averages_eGD = averages_eGD / batch_meta
averages_ts  = averages_ts / batch_meta
```

```python
# Visualize results

plt.figure(figsize=(8,6))
plt.xscale('log')
plt.scatter(meta_runs,result_random, color='k', alpha=0.002)
plt.axhline(y=np.max(prob), color='k')
plt.plot(unique_meta_runs,averages_r, '--k')
plt.ylabel('Conversion')
plt.xlabel('Amount of page views')
plt.title('Performance of the random choice')
plt.ylim([0.0,0.5])
plt.savefig('random.png')
plt.show()
plt.close()
```
